```python
import pandas as pd
import matplotlib.pyplot as plt

distortion = pd.read_csv('k_value_selection.csv')
df = pd.DataFrame(distortion)

plt.plot(df['k'], df['distortions'])
plt.xlabel('K', fontdict = {'fontsize':17})
plt.ylabel('Distortions', fontdict = {'fontsize':17})
plt.savefig('Distortion.png')
plt.show()
```
